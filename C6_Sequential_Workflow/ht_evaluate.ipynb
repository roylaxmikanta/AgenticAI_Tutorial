```python
# 1. Imports
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

# 2. Define State
class WorkflowState(TypedDict):
    input_text: str
    outline: str
    blob: str
    evaluation: str

# 3. Define Nodes
def outline(state: WorkflowState) -> WorkflowState:
    text = state['input_text']
    state['outline'] = f"Outline of: {text}"
    return state

def blob(state: WorkflowState) -> WorkflowState:
    outline_text = state['outline']
    state['blob'] = f"Expanded blob from: {outline_text}"
    return state

def evaluate(state: WorkflowState) -> WorkflowState:
    blob_text = state['blob']
    state['evaluation'] = f"Evaluation result: {blob_text}"
    return state

# 4. Build Graph
graph = StateGraph(WorkflowState)

# Add nodes
graph.add_node('outline', outline)
graph.add_node('blob', blob)
graph.add_node('evaluate', evaluate)

# Add edges: START -> outline -> blob -> evaluate -> END
graph.add_edge(START, 'outline')
graph.add_edge('outline', 'blob')
graph.add_edge('blob', 'evaluate')
graph.add_edge('evaluate', END)

# 5. Compile
workflow = graph.compile()

# 6. Execute
initial_state = {'input_text': 'This is a test document'}
final_state = workflow.invoke(initial_state)
print(final_state)
```
